```python
from dataclasses import dataclass
from sympy import sympify

@dataclass
class CalcEquation:
    name: str
    formula: str

    def expr(self):
        return sympify(self.formula)

    def evaluate(self, values):
        return float(self.expr().subs(values))
```

```python
M_ed = CalcEquation(
    "M_ed",
    "w * L**2 / 8"
)

V_ed = CalcEquation(
    "V_ed",
    "w * L / 2"
)
```

```python
from sympy import symbols

L, w = symbols("L w")

inputs = {
    L: 6,
    w: 25
}

print(M_ed.evaluate(inputs))
```
