```python
from sympy import symbols, Eq, latex, pprint

# Define symbols
x, y = symbols('x y')

# Define equation/expression
expr = x**2 + 3*x + y

# Display equation
print("Expression:")
pprint(expr)

print("\nLaTeX:")
print(latex(expr))

# Substitution values
subs_dict = {x: 5, y: 2}
subs=subs_dict
expr_substituted = expr.subs(subs, simultaneous=True)
print(latex(expr_substituted))

print("\nSubstitutions:")
for var, value in subs_dict.items():
    print(f"{var} = {value}")

# Show substituted expression
expr_sub = expr.subs(subs_dict)

print("\nAfter substitution:")
pprint(expr_sub)

# Evaluate numerically
result = expr_sub.evalf()

print("\nFinal result:")
print(result)
```

```python
def calc_steps(expr, subs):
    original = latex(expr)

    temp_subs = {
        k: symbols(str(v))
        for k, v in subs.items()
    }

    substituted = latex(expr.xreplace(temp_subs))
    result = latex(expr.subs(subs))

    return rf"""
{original}
=
{substituted}
=
{result}
"""
```

```python
expr = x**2 + 3*x + y
subs = {x: 5, y: 2}
calc_steps(expr, subs)
```

```python
from sympy import *
from sympy.parsing.sympy_parser import parse_expr

def substitute_preserve(expr, subs):
    if expr.is_Atom:
        return subs.get(expr, expr)

    return expr.func(
        *(substitute_preserve(arg, subs) for arg in expr.args),
        evaluate=False
    )

def calc_sheet(expr_string, subs):
    # Parse without reordering
    expr = parse_expr(expr_string, evaluate=False)

    original = latex(expr)

    substituted = latex(
        substitute_preserve(
            expr,
            {k: Integer(v) if isinstance(v, int) else v
             for k, v in subs.items()}
        )
    )

    result = latex(expr.subs(subs).evalf())

    return f"""
{original}
=
{substituted}
=
{result}
"""

# User input
x, y = symbols('x y')

print(
    calc_sheet(
        "x**2 + 3*x + y",
        {x: 5, y: 2}
    )
)
```
