```python
import ipywidgets as widgets
from IPython.display import display

from structcalc.calculations.beam import BeamInputs, simply_supported_beam_udl


span = widgets.FloatSlider(
    value=5,
    min=1,
    max=20,
    step=0.1,
    description="Span (m)",
    continuous_update=True,
)

load = widgets.FloatSlider(
    value=20,
    min=1,
    max=100,
    step=1,
    description="UDL (kN/m)",
    continuous_update=True,
)

summary = widgets.HTML()
steps_box = widgets.VBox()


def latex_block(text):
    return widgets.HTMLMath(
        value=rf"""
        <div style="
            font-size:18px;
            padding:8px;
            margin:4px 0;
            border-left:4px solid #888;
            background:#fafafa;
        ">
        $$ {text} $$
        </div>
        """
    )


def update(*args):
    inputs = BeamInputs(
        span=span.value,
        udl=load.value,
    )

    result = simply_supported_beam_udl(inputs)

    summary.value = f"""
    <h3>Results Summary</h3>
    <p><b>Maximum Moment:</b> {result.max_moment_knm:.2f} kNm</p>
    <p><b>Maximum Shear:</b> {result.max_shear_kn:.2f} kN</p>
    """

    step_widgets = []

    for step in result.steps:
        step_widgets.append(
            widgets.VBox([
                widgets.HTML(value=f"<h4>{step.title}</h4>"),
                latex_block(step.formula),
                latex_block(step.substitution),
                widgets.HTML(
                    value=f"""
                    <div style="
                        color:#666;
                        font-size:12px;
                        margin-top:4px;
                    ">
                    <b>Reference:</b> {step.reference}
                    </div>
                    """
                ),
                latex_block(step.result),
                
            ])
        )

    steps_box.children = step_widgets


span.observe(update, names="value")
load.observe(update, names="value")

ui = widgets.VBox([
    widgets.HTML(value="<h2>Simply Supported Beam with UDL</h2>"),
    span,
    load,
    summary,
    widgets.HTML(value="<h3>Calculation Steps</h3>"),
    steps_box,
])

display(ui)

update()
```
