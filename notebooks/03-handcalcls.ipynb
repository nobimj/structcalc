```python
import handcalcs.render

from sympy import symbols, integrate, diff, simplify, Eq, solve, latex
from IPython.display import display, Math
```

```python
x, L, w, E, I = symbols("x L w E I", positive=True)
C1, C2 = symbols("C_1 C_2")
```

```python
M = w*x*(L - x)/2

display(Math(r"M(x) = " + latex(M)))
```

```python
theta_expr = integrate(M/(E*I), x) + C1
y_expr = integrate(theta_expr, x) + C2

display(Math(r"\theta(x) = " + latex(theta_expr)))
display(Math(r"y(x) = " + latex(y_expr)))
```

```python
equations = [
    Eq(y_expr.subs(x, 0), 0),
    Eq(y_expr.subs(x, L), 0)
]

constants = solve(equations, [C1, C2])

theta = simplify(theta_expr.subs(constants))
y = simplify(y_expr.subs(constants))

display(Math(r"\theta(x) = " + latex(theta)))
display(Math(r"y(x) = " + latex(y)))
```

```python
curvature = simplify(diff(y, x, 2))
M_check = simplify(E*I*curvature)

display(Math(r"\frac{d^2y}{dx^2} = " + latex(curvature)))
display(Math(r"EI\frac{d^2y}{dx^2} = " + latex(M_check)))
```
